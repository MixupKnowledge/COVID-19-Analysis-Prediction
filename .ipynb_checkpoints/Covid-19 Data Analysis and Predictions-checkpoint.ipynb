```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('covid_19_india.csv')
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe()
```

```python
df=df[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]
df.columns=['date','state','cured','deaths','confirmed']
```

```python
df
```

```python
today=df[df.date=='2021-08-11']
```

```python
today.head()
```

```python
max_confirm=today.sort_values(by='confirmed',ascending=False)
```

```python
max_confirm
```

```python
topstateconfirmed=max_confirm[0:5]
topstateconfirmed
```

```python
sns.set({'figure.figsize':(15,8)})
sns.barplot(x='state',y='confirmed',data=topstateconfirmed,hue='state')
plt.show()
```

```python
max_death=today.sort_values(by='deaths',ascending=False)
max_death.head()
```

```python
topstatedeath=max_death[0:5]
topstatedeath
```

```python
sns.set({'figure.figsize':(15,8)})
sns.barplot(x='state',y='deaths',data=topstatedeath,hue='state')
plt.show()
```

```python
max_cured=today.sort_values(by='cured',ascending=False)
max_cured.head()
```

```python
topstatecured=max_cured[0:5]
topstatecured
```

```python
sns.set({'figure.figsize':(15,8)})
sns.barplot(x='state',y='cured',data=topstatecured,hue='state')
plt.show()
```
